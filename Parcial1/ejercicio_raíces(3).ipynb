```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(x):
    return 3*(x**5) + 5*(x**4) - (x**3)
```

```python
x = np.linspace(-2, 1, 100)
y = funcion(x)
```

```python
plt.plot(x, y)
plt.axhline(y = 0,color='r')
```

```python
def derivar(f, x, h=1e-6):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def MétodoNewtonR(f, df, xn, itmax=100, precision=1e-6):
    e = 1
    iteracion = 0    
    while e > precision and iteracion < itmax:
        try:
            xn1 = xn -f(xn)/df(f, xn)
            e = np.abs(f(xn)/df(f, xn))
        except ZeroDivisionError:
            print("División por 0")    
        xn = xn1
        iteracion += 1    
    if iteracion == itmax:
        return False
    else:
        return xn
```

```python
def Raíces(x, tolerancia=3):
    Raices = np.array([])
    for i in x:
        Root = MétodoNewtonR(funcion, derivar, i)
        if Root != False:
            raiz = np.round(Root, tolerancia)
            if raiz not in Raices:
                Raices = np.append(Raices, raiz)
    Raices.sort()
    return Raices
```

```python
x = np.linspace(-2, 1, 100)
Raices = Raíces(x)
print(Raices)
```
